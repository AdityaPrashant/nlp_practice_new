```python
import csv
import jsonlines

# Open the CSV file
with open('duly_sample.csv', mode='r') as csv_file:

    # Create a CSV reader object
    csv_reader = csv.DictReader(csv_file)

    # Loop through each row in the CSV file
    for row in csv_reader:

        # Create an empty dictionary to hold the row data
        data = {}

        # Loop through each column in the CSV file
        for column in csv_reader.fieldnames:

            # Add the column data to the dictionary
            data[column] = row[column]

        # Open the JSONL file in append mode
        with jsonlines.open('your_training_data.jsonl', mode='a') as writer:

            # Write the dictionary to the JSONL file
            writer.write(data)
```

```python
import csv
import jsonlines

# Open the CSV file
with open('chatgpt-1.csv', mode='r') as csv_file:

    # Create a CSV reader object
    csv_reader = csv.DictReader(csv_file)

    # Loop through each row in the CSV file
    for row in csv_reader:

        # Create an empty dictionary to hold the row data
        data = {}

        # Loop through each column in the row
        for column, value in row.items():

            # Add the column data to the dictionary
            data[column] = value

        # Open the JSONL file in append mode
        with jsonlines.open('your_training_data.jsonl', mode='a') as writer:

            # Write the dictionary to the JSONL file
            writer.write(data)
```
